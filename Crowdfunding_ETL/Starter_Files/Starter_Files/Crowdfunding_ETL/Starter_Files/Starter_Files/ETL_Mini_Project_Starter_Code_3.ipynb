**Create a Category DataFrame that has the following columns:**

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df.columns
```

```python
# Assign the category and subcategory values to category and subcategory columns.
crowdfunding_info_df[['category', 'sub-category']] = crowdfunding_info_df['category & sub-category'].str.split("/", expand = True)
crowdfunding_info_df

categories = crowdfunding_info_df['category']
subcategories = crowdfunding_info_df['sub-category']
```

```python
# Get the unique categories and subcategories in separate lists.




print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories.unique()))
print(len(subcategories.unique()))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id.
category_ids = np.arange(1, 10) 
subcategory_ids = np.arange(1, 25)
cat_ids = [f'cat{cat}' for cat in category_ids]

# Use a list comprehension to add "subcat" to each subcategory_id.    
scat_ids =  [f'subcat{scat}' for scat in subcategory_ids]
    
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
# 
unique_c = categories.unique()
cat_dict= {
    'category': unique_c,
    'id': cat_ids
    }
category_df = pd.DataFrame(cat_dict)
# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
unique_s= subcategories.unique()
subcat_dict ={
    'id': scat_ids,
    'sub-category': unique_s
}
subcategory_df = pd.DataFrame(subcat_dict)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.

- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.

```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
campaign_df.rename(columns = {
    'blurb':'description', 
    'launched_at': 'launched_date',
    'deadline': 'end_date'
                    }, inplace = True)
campaign_df
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df['goal'] = campaign_df['goal'].astype(float)
campaign_df['pledged']= campaign_df['pledged'].astype(float)
campaign_df
```

```python
# Check the datatypes
campaign_df.dtypes
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt

campaign_df['launched_date'] = pd.to_datetime(campaign_df['launched_date'])
campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'])



```

```python
# # Merge the campaign_df with the category_df on the "category" column and 
# # the subcategory_df on the "subcategory" column.

campaign_merged_df = pd.merge(campaign_df, category_df, on = 'category')
campaign_merged_df.tail(10)

campaign_submerged_df = pd.merge(campaign_merged_df, subcategory_df, on ='sub-category')

campaign_submerged_df.rename(columns ={'id_x':'category_id', 'id_y': 'subcategory_id'}, inplace = True)

campaign_submerged_df
```

```python
# Drop unwanted columns
campaign_cleaned = campaign_submerged_df.drop(columns = ['staff_pick', 'spotlight'])
campaign_cleaned
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

```python
# Convert the "contact_id" column to an int64 data type.
 
contact_info_df_copy_copy['contact_id'] = contact_info_df_copy_copy['contact_id'].astype(int)

```

```python
# Extract the name of the contact and add it to a new column.

name_p = r'[A-Z][a-z]+ [A-Z][a-z]+'

name_list = []

for cone in contact_info_df_copy_copy['contact_info']:
   name_add= re.findall(name_p, cone)
   name_list.append(name_add[0])
name_list

contact_info_df_copy_copy['names'] = name_list


contact_info_df_copy_copy
```

```python
# Extract the email from the contacts and add the values to a new column.
email_p = r'[\w\.-]+@[\w\.-]+'

email_list = []

for con in contact_info_df_copy_copy['contact_info']:
   email_add= re.findall(email_p, con)
   email_list.append(email_add[0])
email_list

contact_info_df_copy_copy['email'] = email_list

contact_info_df_copy_copy
```
